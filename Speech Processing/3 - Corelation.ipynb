## Auto Correlation

```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
d_file = "../dataset/auds/speech.wav"
d,sr = librosa.load(d_file,8000)
```

```python
plt.figure()
t = np.linspace(0,len(d)/sr,len(d))

plt.plot(t,d,lw=1)
plt.xlabel("Time in seconds")
plt.ylabel("Amplitude")

```

```python
r = librosa.autocorrelate(d,max_size=10000)

plt.figure()
plt.xlabel("Lags")
plt.ylabel("Co-efficient")
plt.acorr(r,maxlags=20)

plt.show()
```

## Cross Correlation

```python
d_file = "../dataset/auds/sample1.wav"
d1_file = "../dataset/auds/sample2.wav"

d,sr = librosa.load(d_file,8000)
d1,sr = librosa.load(d1_file,8000)

correlation = np.correlate(d,d1,mode="full")
lags = np.arange(-100,100)

```

```python
def plot(file,title,xlabel,ylabel,correlation_plot=False):
    plt.figure()
    plt.title(title)
    t = np.linspace(0,len(file)/sr,len(file))
    if correlation_plot:
        plt.plot(lags,d)
    else:
        plt.plot(t,file,lw=1)
    plt.grid(True)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot(d,"Speech File 1","Time","Amplitude")
plot(d1,"Sample File 2","Time","Amplitude")

plot(correlation[0:200],"Cross Correlation Plot","Lags","Co-efficient")
```
