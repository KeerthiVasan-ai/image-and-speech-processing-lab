```python
import numpy as np
import librosa
import librosa.core as lc
import numpy as np
import librosa.display as ld

import matplotlib.pyplot as plt
```

```python
path = "../dataset/auds/speech.wav"
y, sr = librosa.load(path,8000)
```

```python
n_fft =  1024

plt.title("Original Speech Signal")
plt.plot(y,label="Original file")
plt.xlabel("Time")
plt.grid(True)
plt.show()
```

```python
def spectogram(file,hop_length,win_length, title):
    mag = np.abs(lc.stft(file,n_fft=n_fft,hop_length=hop_length,win_length=win_length,window="hamming"))
    mag1 = librosa.amplitude_to_db(mag,ref=np.max)
    img = librosa.display.specshow(mag1,sr=sr,x_axis='s',y_axis='linear')
    plt.title(title)
    return img
    
```

```python
plt.colorbar(spectogram(y,10,40,"WideBand Spectogram"),format="%+2.fdB")
plt.show()
```

```python
plt.colorbar(spectogram(y,100,400,"NarrowBand Spectogram"),format="%+2.fdB")
plt.show()
```
