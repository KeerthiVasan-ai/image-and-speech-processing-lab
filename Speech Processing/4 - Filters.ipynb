## High Pass Filter

```python
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt
import librosa
from scipy.signal import butter,filtfilt
```

```python
d_file = "../dataset/auds/set2.wav"
d,sr = librosa.load(d_file,8000)

t = np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Original Speech File")
plt.grid(True)
plt.show()
```

```python
def filter(file,lower_limit,higher_limit,filter_name,title):
    b,a = sg.butter(lower_limit,float(higher_limit) / (sr/2), filter_name)
    d_fill = sg.filtfilt(b,a,d)

    fig,ax = plt.subplots(1,1,figsize=(6,3))

    t = np.linspace(0.,len(file)/sr,len(file))
    ax.plot(t,file,lw=1)
    ax.plot(t,d_fill,lw=1)

    plt.xlabel("Time")
    plt.ylabel(f"{title} Filter")
    plt.grid(True)
    plt.show()
```

```python
filter(d,4,2000,"high","High Pass")
filter(d,4,500,"low","Low Pass")
```
