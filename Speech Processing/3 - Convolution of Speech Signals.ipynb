```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
d_file = "../dataset/auds/set1.wav"
d1_file = "../dataset/auds/set2.wav"
```

```python
d, sr = librosa.load(d_file,8000)
d1, sr = librosa.load(d1_file,8000)
```

```python
y = signal.convolve(d,d1)
```

```python
def plot_audio(file,xlabel,ylabel):
    plt.figure()
    t = np.linspace(0,len(file)/sr,len(file))
    plt.plot(t,file,lw=1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
```

```python
plot_audio(d,"Time","Amplitude")
plot_audio(d1,"Time","Amplitude")
plot_audio(y,"Time","Amplitude")
```
